[Problem 1] Data visualization

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

csv_path = "https://docbase.io/file_attachments/10df533f-9efa-45ee-ab05-966691804a95.csv" # path

np.set_printoptions(suppress=True) # Set prohibition of scientific notation
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
print(fuji[130:140])

# Display the graph. Title and axis label name.
plt.title("Fuji elevation")
plt.xlabel("the number of the point")
plt.ylabel("elevation")
plt.plot(fuji[:,3])  # Enter the variable name of the list in "List name"
plt.show()
```

[Problem 2] Create a function to calculate the gradient at a certain point

```python

# slope = amountOfChangeY / amountOfChangeX

def findGradient(elevation,numberOfPoint):
  return (elevation)/(numberOfPoint)

```

[Problem 3] Create a function to calculate the destination point

```python
# destination point = current point - $ \ alpha $ x current point gradient

def findDestinationPoint(alpha,currentPoint, currentPointGradient):
  #alpha = 0.2 # learning rate

  dest = currentPoint-alpha*currentPointGradient
  return dest
  
```

[Problem 4] Create a function to go down the mountain

```python
# example of gradient descent for a one-dimensional function
from numpy import asarray
from numpy.random import rand

currentPoint = 136
n_iter = 300
# generate an initial point
startElevation = fuji[currentPoint, 3]
print(startElevation,currentPoint)
points = []
alpha = 0.2
# gradient descent algorithm
for i in range(n_iter):
  gradient = findGradient(startElevation,currentPoint)
  currentPoint = findDestinationPoint(alpha,currentPoint,gradient)
  points.append(currentPoint)
```

[Problem 5] Visualization of the descent process

```python
# Display the graph. Title and axis label name.
plt.title("Altitude and slope")
plt.xlabel("the number of the point")
plt.ylabel("altitude")
plt.plot(points)  # Enter the variable name of the list in "List name"
plt.show()
```
